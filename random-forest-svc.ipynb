Creating the heatmap

```python
import seaborn as sns
from matplotlib.pyplot import figure
figure(figsize= (13,12))
sns.heatmap(dataset.corr(),annot= True)
plt.show()
```

```python
dataset["age"]= np.log(dataset.age)
dataset["trtbps"]= np.log(dataset.trtbps)
dataset["chol"]= np.log(dataset.chol)
dataset["thalachh"]= np.log(dataset.thalachh)
y = dataset['cp'].values
X = dataset.iloc[:,:-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
RFclassifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
RFclassifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
y_pred_RF = RFclassifier.predict(X_test)
cm = confusion_matrix(y_test, y_pred_RF)
figure(figsize= (8,6))
sns.heatmap(cm, annot = True)

print('ACCURACY OF RANDOM FOREST:')
print(accuracy_score(y_test, y_pred_RF))
print('Confusion matrix: ')
```

```python
from sklearn.svm import SVC
SVCclassifier = SVC(kernel = 'rbf', random_state = 0)
SVCclassifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
y_pred_SVC = SVCclassifier.predict(X_test)
cm = confusion_matrix(y_test, y_pred_SVC)
figure(figsize= (8,6))
sns.heatmap(cm, annot = True)

print('ACCURACY OF Support Vector Machine:')
print(accuracy_score(y_test, y_pred_SVC))
print('Confusion matrix: ')
```
